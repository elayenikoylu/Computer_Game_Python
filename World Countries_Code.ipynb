Introduction

For the purposes of our research South America has been defined as the continental Americas south of the Darién Gap; a densely forested, watershed and mountainous region between Panama and Colombia. While the majority of countries in South America may share many surface level similarities like language and similar cuisines, the peoples of this continent are far from a cultural monolith. Although many features of contemporary South American culture can be traced back to their Iberian colonizers, Portugal and Spain’s influence only make up part of the eclectic mix of cultures that comprise the continent in the present day.

The earliest signs of humans in South America date back to just 14,000 years ago and were discovered in Southern Chile. These early inhabitants would go on to become the indigenous peoples of South America. Currently, the continent is home to two countries where indigenous peoples constitute the largest ethnic group. These are Peru, where 45% are indigenous and Bolivia, where 62% of people identify as feeling a part of some indigenous group.
By analyzing data by the World Bank, we will try to determine which country is the most representative of the continent as a whole.

```python
##########################  STAGE6  ##########################



#### Creating a histogram for the column. e.g. variable "prenatal_care"

def histGraph(dataCat,vari,Bin):
    #Function to create histogram given paramenters
    
    col=vari
    data_categ=dataCat.copy()

    tit=Nick_Name[col]

    # setting figure size
    fig, ax = plt.subplots(figsize = [8, 5])

    # histogram for AIDS
    sns.distplot(a     = data_categ[col],
                 bins  = Bin,
                 hist  = True,
                 kde   = False,
                 rug   = False,
                 color = 'black')

    # titles and labels
    plt.title(label = f"Distribution of {tit}")
    plt.xlabel(xlabel = f'{col}')
    plt.ylabel(ylabel = 'Frequency')


    # New: These add vertical lines to the code
    plt.axvline(x = data_categ[col].mean(),
                color = 'red')

    plt.axvline(x = data_categ[col].median(),
                color = 'blue')

    # this adds a legend
    plt.legend(labels =  ['mean', 'median'])

    # these compile and display the plot so that it is formatted as expected
    plt.tight_layout()
    plt.show()
    
#Calling histogram for variables: AIDS,CO2,undernourishment,women_parliaments,primary_school_enrollment_rate
#primary_school_completion
histGraph(dataCat=data_health,vari='AIDS',Bin=4)
```

```python
##########################  STAGE8  ##########################

#       
# Run if only you want to create a correlation matrix.    

def heatGraph(dataCat,categ):
    #Function to create heat graph base on parameters 

    data_categ=dataCat.copy()

    data_corr = data_categ.corr(method = 'pearson').round(decimals = 2)

    # specifying plot size (making it bigger)
    fig, ax = plt.subplots(figsize=(12,12))

    # developing a cool heatmap
    sns.heatmap(data       = data_corr, # the correlation matrix
                cmap       = 'Blues',     # changing to COOL colors
                square     = True,          # tightening the layout
                annot      = True,          # should there be numbers in the heatmap
                linecolor  = 'black',       # lines between boxes
                linewidths = 0.5)          # how thick should the lines be?

    plt.tight_layout()
    plt.savefig(fname = f"./corr_{categ}.jpeg")

    # title and displaying the plot
    plt.title(f"""
    Linear Correlation Heatmap for {categ} Features
    """)

    plt.show()
```

```python
##########################  STAGE9_1  ##########################

data_c = data_eco

#################
#Ranking

#This proces rank for each column then take a average for each country
#################

list_des=['primary_school_enrollment_rate','HIV_antiretvir_cov_%',\
          'skilled_births','female_family_workers','male_family_workers','total_family_workers',\
          'female_employment','male_employment','total_employment',\
          'GDP_employment','GDP_energy','GNI','measles_immunization','sanitation','water',\
         'lowest_20pct_income_share','internet_users','life_expectancy','adult_literacy',\
         'mobile_cellular_subscriptions','net_ODA_received','population','prenatal_care',\
         'primary_school_completion','women_parliaments','primary_school_enrollment','self-employed',\
         'trade']

list_asc=['AIDS','adolescent_fertility_rate','CO2','energy','total_fertility_rate','tuberculosis',\
         'maternal_mortality','infant_mortality_rate','poverty_gap','HIV','undernourishment',\
         'malaria','tuberculosis_death','AIDS_ratio']

def ranking(data, listdes, listasc, MaxNAs):
    
    #Those following three lines create a partially empty dataframe in which the code
    #will fill the ranking of each country by its column.
    data_t=pd.DataFrame.copy(data[['Country Code Total','Country Name','Hult Region','Cool Name']])
    temp = pd.DataFrame(columns=list(data.columns)[4:], index=range(len(data)))
    data_t=pd.concat([data_t,temp],axis=1)
    
    #This line makes a data frame to store the average ranking for each country.
    data_rank=pd.DataFrame(columns=['Country','Ranking Average','Range of Notch'], index=range(len(data)))
    
    
    temp=list(temp.columns)
    
    #These loops will sort and set the ranking for each column. 
    for i in temp:
        obs=len(data)
        
        #The following conditional set which kind of sort the column 'i' should have.
        #For example, if i=GNI, the sort is descending (binary=False). Or if i=CO2, binary=True (ascending).
        binary=True
        if i in listdes:
            binary=False
        elif i in listasc:
            binary=True
        else:
            print('There is a mistake')
            break
        
        #This line sorts the values of the collumn 'i' and locates in which row of the data_t
        #is the country with the better value of 'i'. 
        Ix=data[['Country Name',i]].sort_values(by=i, ascending=binary,axis=0).index
        
        ranks=len(data)-data[i].isnull().sum(axis=0)
                
        for j in range(ranks):
            data_t.loc[Ix[j],i]=obs
            obs -=1
        
    for i in range(len(data)):
        
        data_rank.loc[i,'Country']=data_t.loc[i,'Country Name']
        
        #If the country has few missing values (below the MaxNAs), it will be able
        #to be part of the ranking.
        if data_t.iloc[i,4:].isnull().sum(axis=0)<=MaxNAs: 
            
            row_t=pd.DataFrame(data_t.iloc[i,4:]).iloc[:,0]
            mean_t=row_t.mean().round(2)
            min_t=row_t.min()
            max_t=row_t.max()
            
            #data_rank.loc[i,'Country']=data_t.loc[i,'Country Name']
            data_rank.loc[i,'Ranking Average']=mean_t
            data_rank.loc[i,'Range of Notch']=max_t-min_t  
        
    data_rank.sort_values(by='Ranking Average',ascending=False,axis=0, inplace=True)
   
    return [data_t,data_rank]

rank_health=ranking(data=data_health, listdes=list_des, listasc=list_asc,MaxNAs=4)

rank_health[1]
```

```python
##########################  STAGE10  ##########################

#### Creating a histograms

data_categ=data_eco.copy()

# setting figure size
fig, ax = plt.subplots(figsize = [20, 20])

def histMult(col,place,bi,rows,cols):
    #This function plots multiple histograms
    #col=the serie, place=which part of the histogram frame is
    #bi=number of bins and rows and cols the size of the figure.

    tit=Nick_Name[col]

    plt.subplot(rows, cols, place)
    sns.distplot(a    = data_categ[col],
                 bins = bi, 
                 hist = True, 
                 kde  = False, 
                 rug  = False, 
                 color='black')

    # titles and labels
    plt.title(label = f"Distribution of {tit}")
    plt.xlabel(xlabel = f'{col}')
    plt.ylabel(ylabel = 'Frequency')

    # New: These add vertical lines to the code
    plt.axvline(x = data_categ[col].mean(),color = 'red')
    plt.axvline(x = data_categ[col].median(),color = 'blue')

    # this adds a legend
    plt.legend(labels =  ['mean', 'median'])

#Calling histogram for each serie or colum
lst=['total_family_workers','total_employment','GDP_employment','GDP_energy',
    'GNI','lowest_20pct_income_share','net_ODA_received','self-employed','trade']

count=1
for i in lst:
    
    histMult(col=i,place=count,bi=10,rows=3,cols=3)
    count +=1

del [i, count, lst]

# these compile and display the plot so that it is formatted as expected
plt.tight_layout()
plt.savefig(fname = './hist_eco.jpeg')

plt.show()
```

```python
##########################  STAGE10  ##########################

#### Creating boxplots

data_categ=data_eco.copy()

# setting figure size
fig, ax = plt.subplots(figsize = [15, 35])

def boxplotMult(Xs,Dat,place,rows,cols,Ys=None,Hue=None,Orient='h'):
    #This function plots multiple histograms
    #col=the serie, place=which part of the histogram frame is
    #bi=number of bins and rows and cols the size of the figure.

    tit=Nick_Name[Xs]
    if Hue!=None:per=Nick_Name[Hue]
        
    plt.subplot(rows, cols, place)
    
    sns.boxplot(x      = Xs,   # x-variable
                y      = Ys, # optional y-variable
                hue    = Hue,      # optional categorical feature
                orient = Orient,       # horizontal or vertical
                data   = Dat)  # DataFrame where features exist

    
    # titles and labels
    if Hue==None:
        plt.title(label   = f"""
    Outlier Detection:
    Boxplots for {tit}
    """)
    else:
        plt.title(label   = f"""
    Outlier Detection:
    Boxplots for {tit} per {per}
    """)             
        
    plt.xlabel(xlabel = f'{Xs}')
    plt.ylabel(ylabel = f'{Ys}')

    # New: These add vertical lines to the code
    plt.axvline(x = data_categ[Xs].mean(),color = 'red',label='mean')
    plt.axvline(x = data_categ[Xs].median(),color = 'black',label='median')

    # this adds a legend
    plt.legend()

#Calling histogram for each serie or colum
lst=['total_family_workers','total_employment','GDP_employment','GDP_energy',
    'GNI','lowest_20pct_income_share','net_ODA_received','self-employed','trade']

count=1
for i in lst:
    boxplotMult(Xs=i,Dat=data_categ,place=count,rows=9,cols=2,Ys=None,Hue=None,Orient='h')
    count +=1
    boxplotMult(Xs=i,Dat=data_categ,place=count,rows=9,cols=2,Ys='Country Name',Hue=None,Orient='h')
    count +=1
    
del [i, count, lst]

# these compile and display the plot so that it is formatted as expected
plt.tight_layout()
plt.savefig(fname = './boxplot_eco.jpeg')

plt.show()
```
