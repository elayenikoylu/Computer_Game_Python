```python
"""
Author: @ela.yenikoylu

DocString:
    A) Introduction:
    This is the space game championship // First Human Trip to Mars. This game
    Includes total 4 rounds:
    
    Game Start Function
    Round 1: Truth or Dare
    Round 2: Truth Questions
    Round 3: Pick a spaceship
    Round 4: Additional Round Kongo Spaceship Training 
    Countdown Function
    Fail Function
    Game End
    
 """
#Importing necessary functions for our code.


from random import randint
import random
import time

#In this function we are letting users play a little truth or dare to make a 
#start to our space game.

def start():
    
    global gamer_name # Calling gamer_name as global becasue I will use it in 
    # most of the different functions below.
    
    print('''\nHello and welcome to the Space Game Championship 'First Human Trip 
to Mars'\n''')

    input('<Press any key to continue>\n')
    
    print("My name is Ela and I will assist you throughout the entire game!", end = " ")

    gamer_name = input(prompt = 'What is your name? \n')
    gamer_name = gamer_name.capitalize()
    
    print(f"\nWelcome {gamer_name}!!!!\n") # Welcoming the users
    input(prompt = '<Press any key to continue>\n')
    truth_dare() # Caliing the truth_dare function for the next steps.

    
# This functions is created to ask users play a Truth or Dare game.    
    
def truth_dare():
    
    print("""
Ela : In this round we will play a little truth or dare!

Let's get started!!!""", end = " ")
    
    # Asking users truth or dare.
    
    gamer_answer = input(prompt = 'Please pick: Truth or Dare?\n') 
    
    # Conditional statements to evalueate inputs for the question.

    if gamer_answer.lower() == "truth":
        print(f'''
Ela: Alright {gamer_name}, you picked truth! Get ready for your question. 
You will need 2 correct answers to pass the round and be careful, if you picked 
the wrong answer 3 times, you will lose the game!''', end = " ")
        truth() # Calling the truth funtion to go next steps with the questions.
           
        
    elif gamer_answer.lower() == "dare":
        print(f'''Woooww ,{gamer_name} impressive! As an award for being so brave, 
you just passed this round!''', end = " ")
        pick_spaceship() # Calling pick_spaceship function if the input is "dare".
           
            
    else:
        print("Something went wrong. Please input 'Truth' or 'Dare'", end = " ")
        truth_dare() # If there is an answer different than "truth" or "dare" 
        #takes users to the beginning of the same function.
      
  
  
# This function is for users picked 'Dare'. They will answer questions to go 
# to the next level.

def truth():
    
    # Creating two different variables with same content to use it in the while 
    # loop below to avoid repating the same questions in the loop.
    
    questions_truth = ['What color is the Mars?',
                      'How many months would it take to travel to Mars?',
                      'Which one is not one of the moons Mars have?',
                      'How far is mars from the Sun? (in million mi)']
    
    questions_pick  = ['What color is the Mars?',
                      'How many months would it take to travel to Mars?',
                      'Which one is not one of the moons Mars have?',
                      'How far is mars from the Sun? (in million mi)']
  
   
    times_fail = 3
    times_correct = 2
    unknown_input = 3
    
# While loop below asks random questions from the above list (questions_pick) to users.

    while times_fail > 0 and times_correct > 0 and unknown_input > 0 : 
        question = random.choice(questions_pick)
        questions_pick.remove(question)
        input(prompt  = "<Press enter for your question>\n")
        print(question)
        print(" ")
        
# Creating if/elif/else statements within the loop to design our code based on 
#answeres users pick.   

        # I used the other list (questions_truth) to avoid indexing errors.
        if question == questions_truth[0]: 
            print("1) Red")
            print("2) White")
            print("3) Yellow")
                    
            color = input(prompt = '> \n')
            
            if color == "1" or color.lower() == "red":
                times_correct -= 1
                if times_correct == 0 :
                    print(f'''
Ela: Good job {gamer_name}, you passed this round!''', end = " ")
                    pick_spaceship() # Sending users to another level 
                    break            # if they give 2 correct answers.
                
                elif times_correct >0:
                    print(f'''
Ela: Good job {gamer_name}, you need one more correct answer to pass the round.''', end = " ")
                
            
          
            elif color == "2" or color.lower() == "white" or color == "3"\
        or color.lower == "yellow":
                times_fail -= 1
                
                print(f'''
Ela: Oh {gamer_name}, I am sorry but that's incorrect! If you fail {times_fail} 
time(s) more you will lose the game ''', end = " ")
                
                if times_fail == 0:
                    print('''
Ela: You answered 3 quesitons wrong!''', end = " ")
                    fail()  # Sending users to fail function if they give 3 wrong answers.
                    break
                
# If there is more than 3 unknown inputs, failing the users with a warning message.
            else:  
                unknown_input -= 1
                print('''Something went wrong.''', end = " ")
                if unknown_input == 0:
                    print (f''' Please take the game seriously and follow the instructions''')
                    fail()
                
                
        
        elif question == questions_truth[1]: # Same structure for another question.
            print("1) 5")
            print("2) 9")
                    
            months = input(prompt = '> \n')
            
            if months == "2" or months == "9":
                times_correct -= 1
                
                if times_correct == 0:
                    print(f'''
Ela: Good job {gamer_name}, you passed this round!''', end = " ")
                    pick_spaceship()
                    break
                
                elif times_correct > 0:
                    print(f'''
Ela: Good job {gamer_name}, you need one more correct answer to pass the round.''', end = " ")
                
                
                  
            elif months == "1" or months == "5":
                times_fail -= 1
                print(f'''
Ela: Oh {gamer_name}, I am sorry but that's incorrect! If you fail {times_fail} 
time(s) more you will lose the game ''', end = " ")              
                
                if times_fail == 0:
                    print('''
Ela: You answered 3 quesitons wrong!''', end = " ")
                    fail()  
                    break
            
            else:
                unknown_input -= 1
                print('''Something went wrong.''', end = " ")
                
                if unknown_input == 0:
                    print (f'''Please take the game seriously and follow the instructions''')
                    fail()
              
                
                              
        elif question == questions_truth[2]: # Same structure for another question.
            print("1) Phobos")
            print("2) Deimos")
            print("3) Calios")        

            moons = input(prompt = '> \n')
                    
            if moons == "1" or moons == "2" or moons.lower() == "phobos"\
            or moons.lower() == "deimos":
                times_fail -= 1
                
                print(f'''
Ela: Oh, I am sorry {gamer_name} but that's incorrect! If you fail {times_fail} 
time(s) more you will lose the game''', end = " ")
                
                if times_fail == 0:
                    print('''
Ela: You answered 3 quesitons wrong!''', end = " ")
                    fail()  
                    break
               
                
            elif moons == "3" or moons.lower() == "calios":
                times_correct -= 1
                
                if times_correct == 0:
                    print(f'''
Ela: Good job {gamer_name}, you passed this round!''', end = " ")
                    pick_spaceship()
                    break
                
                elif times_correct >0:
                    print(f'''
Ela: Good job {gamer_name}, you need one more correct answer to pass the round.''', end = " ")
                
            else:
                unknown_input -= 1
                print('''Something went wrong.''', end = " ")
                if unknown_input == 0:
                    print (f''' Please take the game seriously and follow the instructions''')
                    fail()
                               
                
        elif question == questions_truth[3]: # Same structure for another question.
            print("1) 132")
            print("2) 256")
            print("3) 182") 
            
            distance = input(prompt = '> \n')
            
            if distance == "2" or distance == "3" or distance == "256"\
            or distance == "182":
                times_fail -= 1
                print(f'''
Ela: Oh, I am sorry {gamer_name} but that's incorrect! If you fail {times_fail} 
time(s) more you will lose the game''', end = " ")
                if times_fail == 0:
                    print('''
Ela: You answered 3 quesitons wrong!''', end = " ")
                    fail()  
                    break
               
                
            elif distance == "1" or distance == "132":
                times_correct -= 1
                
                if times_correct == 0:
                    print(f'''
Ela: Good job {gamer_name}, you passed this round!''', end = " ")
                    pick_spaceship()
                    break
                
                elif times_correct > 0:
                    print(f'''
Ela: Good job {gamer_name}, you need one more correct answer to pass the round.''', end = " ")
                
            else:
                unknown_input -= 1
                print('''Something went wrong.''', end = " ")
                if unknown_input == 0:
                    print (f'''Please take the game seriously and follow the instructions''')
                    fail()
                
                
# This is round 3. Users made it here after first two rounds, will need 
# to pick a spaceship. 

def pick_spaceship():
    
    print(f'''
Ela: Alright, congrats on the Truth or Dare!!! 

It's time to pick your spaceship. Please pick one of the following:

1) Spaceship Merlot
2) Spaceship Dream
3) Spaceship Kongo

Use your gut feeling and be aware that: One of the ships will give you a VIP trip. 
Another one of them will get you another additional process before you can win. The 
last one is a fake ship, if you chose it you will lose the game :)''')

    choice = input(prompt = '> \n')
    
    # Basic conditional statements to evaluate the inputs.
    
    if choice == "1" or "merlot" in choice.lower():
        print(f'''
Ela: Congrats {gamer_name}, you won!! You got to be a VIP passenger for 
the first Mars trip. Let's get ready for the countdown!''')
        countdown() #Winning users goes to the rocket countdown.
        
         
    elif choice == "2" or "dream" in choice.lower():
        print(f'''
Ela: Oh no {gamer_name}, you picked the fake ship. Forget about 
the Mars!''', end = " ")
        fail() # Sending users to fail function.
        
    elif choice == "3" or "kongo" in choice.lower():
        print(f'''
Ela: Well.... {gamer_name}, you might be able to go to Mars with 
this ship but you need additional training!''', end = " ")
        additional_training() # Sending users for the additional process.
        
    else:
        print('''Something went wrong. Please follow the instructions!''')
        pick_spaceship() # If there is an unknown input, sending the users 
        # at the beginning of the function.

        
# This additional round is for users picked spaceship Kongo. 

def additional_training():
    print(f'''
Ela : Alright, you came a long way! In the ship Kongo, 
the team is a lot more sensitive so to be able to join them you need to 
get some training done. Let's follow the instructions:
    
Let's assume you have arrived to Mars. You just landed. What would be the 
first thing you do?:

1) 'Jump' out of the ship and explore
2) 'Wait' for captain to guide you
3) 'Ask' questions to find out what to do
    
    *** Please refer your answer with the words 'jump', 'wait', and 'ask' or 1,2,3!
    
    ''')
    answer = input(prompt = '> \n')
    
    # Basic conditional statements.
    
    if answer == "1" or answer.lower() == "jump":
        print(f'''
Ela: Oh, no {gamer_name}, that would create a big caos and the 
crew of Kongo would't like that!''', end = " ")
        fail()
    
    elif answer == "2" or answer.lower() == "wait":
        print(f'''
Ela: Well done {gamer_name}, the crew of Kongo would enjoy having 
you on board. Let's get ready for the countdown!\n''')
        countdown()
       
    
    elif answer == "3" or answer.lower() == "ask":
        print(f'''
Ela: That's an alright approach here {gamer_name}, the crew had some 
dicussion and they decide having you on board. Congrats!''', end = " ")
        
    else:
        print('''Something went wrong. Please follow the instructions!''')
        additional_training() # If there is an unknown input, sending the users 
        # at the beginning of the function.
    
# Countdown function for the users won the game.

def countdown():
    
    count = 10
    while count > 0:
        print(count)
        count = count - 1
        time.sleep(1.0)
    print("Enjoy your trip to Mars!", end = " ")
    
    
#Fail function for users failed in any step of the game.

def fail():
    print(f'''{gamer_name}, that's unfortunate but you just lost the game!''', end = " ")
                
    again = input(prompt = '''Would you like to play again? Yes or No? \n''')
             
    if again.lower() == "yes":
        truth_dare()
   
    else:
        print(f'''Thanks for playing {gamer_name}!''')
        
          
# Calling the first function to run the code.            
start()       

#### Game Over :)
```
